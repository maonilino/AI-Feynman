```python

import os
import random

def getY(x01,x02,x03,x04,x05,x06,x07,x08,x09,x10):
  y = 2.5 * ((x01+x02+x03+x04+x05+x06+x07+x08+x09+x10)/10)
  return y

def getRow():
  [x0,x1,x2,x3,x4,x5,x6,x7,x8,x9]=[random.uniform(10,30) for x in range(10)]
  y=getY(x0,x1,x2,x3,x4,x5,x6,x7,x8,x9)
  return str(x0)+" "+str(x1)+" "+str(x2)+" "+str(x3)+" "+str(x4)+" "+str(x5)+" "+str(x6)+" "+str(x7)+" "+str(x8)+" "+str(x9)+" "+str(y)+"\n"

with open("F1.txt", "w") as f:
  for _ in range(10000):
    f.write(getRow())
f.close()
```

```python
import aifeynman

aifeynman.run_aifeynman("./", "F1.txt", 60, "14ops.txt", polyfit_deg=3, NN_epochs=100)
```
